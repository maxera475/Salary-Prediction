```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/job_placement.csv')
```

```python
df.sample(15)
```

```python
df.isnull().sum()
```

```python
df.drop(['id','name','degree','college_name'],axis=1,inplace=True)
```

```python
df['stream'].value_counts()
```

```python
df
```

```python
df1=df.drop(['salary'],axis=1)
```

```python
df1
```

```python
x_train , x_test , y_train , y_test = train_test_split(df1,df.iloc[:,-3],test_size=0.25)
```

```python
x_train.head()
```

```python
si = SimpleImputer()
x_train_years_of_experience= si.fit_transform(x_train[['years_of_experience']])
x_test_years_of_experience= si.fit_transform(x_test[['years_of_experience']])
```

```python
oe = OneHotEncoder(drop='first',sparse=False)
x_train_new = oe.fit_transform(x_train[['gender','stream','placement_status']])
x_test_new = oe.transform(x_test[['gender','stream','placement_status']])
```

```python
x_train_rem = x_train.drop(columns=['years_of_experience','gender','stream','placement_status'],axis=1)
x_test_rem = x_test.drop(columns=['years_of_experience','gender','stream','placement_status'],axis=1)
```

```python
x_train_transformed = np.concatenate((x_train_rem,x_train_years_of_experience,x_train_new),axis=1)
x_test_transformed = np.concatenate((x_test_rem,x_test_years_of_experience,x_test_new),axis=1)
```

```python
x_train_transformed.shape
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(x_train_transformed, y_train)

y_pred = logreg.predict(x_test_transformed)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed,y_train)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train_transformed,y_train)
```

```python
y_pred1 = reg.predict(x_test_transformed)
```

```python
y_pred2 = clf.predict(x_test_transformed)
```
